```python
N = 10 ** 7
n_die = 2

die_tosses = np.random.randint(1, 7, (N, n_die))
results = die_tosses.sum(axis = 1)
unique, counts = np.unique(results, return_counts=True)
rel_counts = counts / np.sum(counts)
plt.plot(unique, rel_counts);
```

```python
plt.bar(unique, rel_counts);
```

```python
print(f"Mean: {np.mean(results)}")
print(f"Variance: {np.var(results)}")
print(f"Median: {np.median(results)}")
```
