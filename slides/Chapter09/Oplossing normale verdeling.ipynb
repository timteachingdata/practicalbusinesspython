```python
import numpy as np
import matplotlib.pyplot as plt
a = 1
```

```python
def normal_bounded(loc, scale, lb, ub):
    
    rand_num = np.random.normal(loc = loc, scale = scale)
    while (rand_num < lb) or (rand_num > ub):
        rand_num = np.random.normal(loc= loc, scale = scale)
    return rand_num
```

```python
N = 10 ** 5
loc = 1 # mean of normal distribution
scale = 0.1 # std of normal distribution
lb = 0.9 # interval: [lb, ub]
ub = 4

data = np.array([normal_bounded(loc, scale, lb, ub)
                                     for _ in range(N)])
```

```python
plt.hist(data, bins = int(np.sqrt(N)) // 2);
```
