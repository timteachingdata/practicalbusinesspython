## Pandas Series

### What is a `pd.Series` object?

```python
data = pd.Series([0.25, 0.5, 0.75, 1])
display(data)
print(data.index)
print(data.values)
print(data.dtype)
```

### Using `.items()`

```python
for idx, item in data.items():
    print(f"{idx} -- {item}")
```

### Alternative indexing

```python
data = pd.Series([0.25, 0.5, 0.75, 1.0],
                    index = ['a', 'b', 'c', 'd'])

print(data['a'])
```

### Series from a dictionary

```python
population_dict = {
                        'California' : 38332521,
                        'Texas' : 26448193,
                        'New York' : 19651127,
                        'Florida' : 19552860,
                        'Illinois' : 12882135
                    }
population = pd.Series(population_dict)
population['California' : 'New York']
```

## Pandas DataFrames

### Creating Pandas DataFrames

#### From Pandas Series

```python
population_dict = {
                        'California' : 38332521,
                        'Texas' : 26448193,
                        'New York' : 19651127,
                        'Florida' : 19552860,
                        'Illinois' : 12882135
                    }
population = pd.Series(population_dict)

area_dict = {
                    'California' : 423967,
                    'Texas' : 695662,
                    'New York' : 141297,
                    'Florida' : 170312,
                    'Illinois' : 149995
                }
area = pd.Series(area_dict)

df_states = pd.DataFrame({
                            'population' : population,
                            'area' : area
                        })

display(df_states)
```

### Indexen zijn verzamelingen met volgorde

## pd.Series indexeren

```python
data = pd.Series(np.arange(0.25, 1.3, 0.25),
                    index = ['a', 'b', 'c', 'd', 'e'])

print(data['e'])
print(data['a':'c'])
print(data[0:2])
print(data[(data > 0.3) & (data < 0.8)])
print(data[['a', 'e']])
```

```python
data = pd.Series(['a', 'b', 'c'], index = [1, 3, 5])
print(data[1])
print(data[1:3])
print(data.loc[1])
print(data.loc[1:3])
print(data.iloc[1])
print(data.iloc[1:3])
```
